```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler

# Loading the dataset
iris_dataset = datasets.load_iris()
# Q 3.1 Using sklearn's StandardScaler to fit and transform the data
scaler = StandardScaler()
iris_dataset['data'] = scaler.fit_transform(iris_dataset['data']) # It removes the mean and scales it to unit variance

iris_dataset
```

```python
# Q 3.2 
from sklearn.preprocessing import LabelEncoder

# Printing to check if target names is getting changed
print(iris_dataset['target_names'])

# Creating LabelEncoder instance and fit_transform to fit and transform at the same time
label_encoder = LabelEncoder()
iris_dataset['target_names'] = label_encoder.fit_transform(iris_dataset['target_names']) 
# It changes the target_names to numerical values ([0, 1, 2]) instead of strings ['setosa' 'versicolor' 'virginica']
iris_dataset['target_names']
```
