```python
from sklearn import datasets
from sklearn.decomposition import PCA

# Loading iris dataset
iris_dataset = datasets.load_iris()

# Performing principal component analysis and reducing to 2 components
pca = PCA(n_components=2)
iris_dataset['data'] = pca.fit_transform(iris_dataset['data'])
iris_dataset
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt

data = iris_dataset['data']
species = iris_dataset['target']

# Scatter plot with different colors for each species
for i, target_name in enumerate(iris_dataset['target_names']):
    plt.scatter(data[species == i, 0], data[species == i, 1], label=target_name) # Taking x and y for each species

# Adding labels, title, grid
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.grid()
plt.legend()
```
