```python
from sklearn import datasets
import numpy as np
import pandas as pd

# Loading as a DataFrame
iris_dataset = pd.DataFrame(datasets.load_iris()["data"])
# Renaming the columns that get saved as 0, 1, 2, 3
iris_dataset.columns = ['sepal length(cm)', 'sepal width(cm)', 'petal length(cm)', 'petal width(cm)']
# Q2.1 a)
print(iris_dataset.describe())
```

```python
# Q2.1 b)
iris_dataset.isnull().sum()

# We can see there are no null columns so we do not need to handle it
```
